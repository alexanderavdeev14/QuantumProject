```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 2
psi0 = basis(N,0) + basis(N,1)
wl = 1
delta_w = 0.15
H = wl* 0.5*sigmaz()
C_ops = [delta_w *sigmaz()]
times = np.linspace(0,50,400)

result = mesolve(H, psi0, times, C_ops)
#result.states

plt.plot(times,np.real(expect(result.states, projection(2,1,0))))
plt.plot(times,np.imag(expect(result.states, projection(2,1,0))))

plt.plot(times,np.real(expect(result.states, projection(2,0,1))))
plt.plot(times,np.imag(expect(result.states, projection(2,0,1))))
#The plot look like ramsy experement



#fig, ax = plt.subplots()
#ax.plot(times,result.states)
#plot.show()
#plt.plot(times,result.expect);
#plt.show()
```
