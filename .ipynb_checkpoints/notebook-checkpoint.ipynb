## Introduction
In this project we will explore different types of dephasing of the 1/2 spin qubit: pure dephasing, dephasing due to decay and depahsing due to termal decoherence. 

##  Description of the physical problem we want to simulate
In the project we will simulate three different experements.  

$\rho$  = 
\begin{matrix}
\rho_{11} \ \rho_{12}\\
\rho_{21} \ \rho_{22}
\end{matrix}

The first experement is the pure dephasing. For this type of depasing T_2 $\neq$ 2T_1.

```python
N = 2
psi0 = basis(N,0) + basis(N,1)
wl = 1
delta_w = 0.15
H = wl* 0.5*sigmaz()
C_ops = [delta_w *sigmaz()]
times = np.linspace(0,50,400)

result = mesolve(H, psi0, times, C_ops)
#result.states

plt.figure()
plt.subplot(211)
plt.plot(times,np.real(expect(result.states, projection(2,1,0))), label=r'Re($\rho_{12}$)');
plt.plot(times,np.imag(expect(result.states, projection(2,1,0))), label=r'Im($\rho_{12}$)' );
plt.legend()

plt.subplot(212)
plt.plot(times,np.real(expect(result.states, projection(2,0,1))), label=r'Re($\rho_{21}$)');
plt.plot(times,np.imag(expect(result.states, projection(2,0,1))), label=r'Im($\rho_{21}$)');
plt.legend()
#The plot look like ramsy experement



#fig, ax = plt.subplots()
#ax.plot(times,result.states)
#plot.show()
#plt.plot(times,result.expect);
#plt.show()
```
